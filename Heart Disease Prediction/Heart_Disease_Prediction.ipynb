Import the libraries


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Import the data

```python
# Load the data into the pandas dataframe
heart_data = pd.read_csv(f'/content/heart_disease_data.csv')
heart_data.head()
```

**Exploratory Data Analysis**

```python
# Let us get some basic information about the data
heart_data.info()
```

```python
# Let us see whether we have any missing values or not in the columns
heart_data.isna().any()
```

```python
# Let us check whether we have any missing values or not in row
heart_data.isna().any(axis=1)
```

```python
# Let us check the number of rows and columns
heart_data.shape
```

```python
# Let us see the statistical information about the data
heart_data.describe()
```

```python
# Let us see the correlation between the features
corr = heart_data.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
# Let us check whether data is biased or not
heart_data['target'].value_counts()
```

The data set is not biased

Let us split the data into predictor and target variables.

```python
X = heart_data.drop(columns = 'target')
y = heart_data['target']
print("Shape of X is ", X.shape)
print("Shape of y is ", y.shape)
```

Let us split the data set into training and testing set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, train_size = 0.8, random_state = 42)
print("Shape of X_train is ", X_train.shape)
print("Shape of y_train is ", y_train.shape)
print("Shape of X_test is ", X_test.shape)
print("Shape of y_test is ", y_test.shape)
```

Train the model

```python
trained_model = LogisticRegression()
```

```python
trained_model.fit(X_train, y_train)
```

Evaluation of the model

```python
# accuracy on the training data
X_train_pred = trained_model.predict(X_train)
acc_score_train = accuracy_score(y_train, X_train_pred)
print("The accuracy on the training set is ", acc_score_train)
```

```python
# accuracy on the testing data
y_pred = trained_model.predict(X_test)
acc_score_test = accuracy_score(y_test, y_pred)
print("The accuracy on the training set is ", acc_score_test)
```

Build a Predictive System

```python
input_data = (60,1,0,130,206,0,0,132,1,2.4,1,2,3)

# change the input array to numpy array
input_data_as_np = np.asarray(input_data)

# reshape the np array
input_data_reshaped = input_data_as_np.reshape(1,-1)

pred = trained_model.predict(input_data_reshaped)
if pred == 0:
  print("Person does not have a heart disease")
else:
  print("Person has a heart disease")
```
